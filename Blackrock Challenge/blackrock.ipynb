```python
import pandas as pd
import os
import numpy as np
from scipy.optimize import minimize
```

```python
# Directory containing CSV files
csv_directory = r'C:\Users\fabro\OneDrive - Imperial College London\1. Quant Projects\Algothon Challenges\Challenge 1\universe'  # Replace with your directory path

# Create an empty list to store DataFrames
dfs = []

# Iterate through the files in the directory
for filename in os.listdir(csv_directory):
    if filename.endswith('.csv'):
        file_path = os.path.join(csv_directory, filename)
        df = pd.read_csv(file_path)
        df.set_index('Date', inplace=True)
        ticker = os.path.splitext(filename)[0]  # Extract the ticker from the filename
        df.columns = [ticker]  # Add a 'ticker' column with the ticker name
        dfs.append(df)

# Concatenate the DataFrames in the list
merged_df = pd.concat(dfs, axis=1,join='inner')  
```
